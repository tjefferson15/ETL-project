```python
happiness19 = "/Users/tjeff/Desktop/CODING/ETL-project/Resources/2019.csv"
df2019 = pd.read_csv(happiness19)
df2019.head()
```

```python
def rounding (x):
    return round (x,2) 
```

```python
drop2016 = df2016.drop(columns = ['Region','Lower Confidence Interval', 'Upper Confidence Interval', 'Economy (GDP per Capita)','Family','Freedom','Dystopia Residual'])
drop2016.head()
```

```python
df2019.columns
```

```python
drop2019 = df2019.drop(columns = ['Overall rank','GDP per capita', 'Social support','Freedom to make life choices']) 
drop2019.head()
```

```python
drop2016_change = drop2016.rename(columns={"Country": "country", 
                                                      "Happiness Rank": "happiness_rank_2016", 
                                                      "Happiness Score": "happiness_score_2016",
                                                      "Health (Life Expectancy)": "life_expectancy_2016",
                                                      "Generosity": "generosity_2016", "Trust (Government Corruption)": "trust_2016"})


drop2016_order = drop2016_change[["country", "happiness_rank_2016", "happiness_score_2016", "life_expectancy_2016", "generosity_2016", "trust_2016"]]
drop2016.head()
```

```python
drop2016_round = drop2016_order
drop2016_round[["happiness_score_2016", "life_expectancy_2016", "generosity_2016", "trust_2016"]] = drop2016_round[["happiness_score_2016", "life_expectancy_2016", "generosity_2016", "trust_2016"]].apply(rounding)
drop2016_round.head()
```

```python
drop2019_change = drop2019.rename(columns={"Country or region": "country", 
                                                      "Score": "happiness_score_2019",
                                                      "Healthy life expectancy": "life_expectancy_2019",
                                                      "Generosity": "generosity_2019",
                                                      "Perceptions of corruption": "trust_2019"})
drop2019_change.head()
```

```python
drop2019_round = drop2019_change
drop2019_round[["happiness_score_2019", "life_expectancy_2019", "generosity_2019", "trust_2019"]] = drop2019_round[["happiness_score_2019", "life_expectancy_2019", "generosity_2019", "trust_2019"]].apply(rounding)
drop2019_round.head()
```

```python
merge_df =pd.merge(drop2016_order, drop2019_change, on="country", how="outer")
merge_df.head(155)
```
